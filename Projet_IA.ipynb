```python
import json
import pandas as pd

# Lecture du fichier JSON
data = []

with open('moderation.jsonl', 'r') as file:
  for line in file:
    data.append(json.loads(line))

# Conversion des données en dataFrame Pandas pour faciliter la manipulation
df = pd.DataFrame(data)
print(df.head())
```
